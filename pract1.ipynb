Цель работы: Провести разведочный анализ данных по датасету. Составить графики, проанализировать их.

```python
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sb

df = pd.read_csv('most_visited_nps_species_data.csv')
df
```

Здесь были импортированы нужные библиотеки и загружен датасет с помощью библиотеки Pandas.

```python
df_new = df.drop(columns=['Synonyms','ExternalLinks','StateStatus','ParkTags','OzoneSensitiveStatus','NativenessTags','OccurrenceTags','Occurrence','Nativeness','GRank','SRank','RecordStatus'])
df_new
```

Здесь с помощью библиотеки pandas был создан новый датафрейм из датасета, а также удалены из него некоторые столбцы, которые были пустыми либо ненужными.

```python
#df_new = df_new.dropna()
df_new['Sensitive'] = df_new['Sensitive'].replace(False, True)
df_new
```

Были заменены значения в столбце Sensitive с False на True, а также показан метод для удаления значений NaN (не использован так как таких значений очень много).

```python
df_new.sort_values(by='SciName')
```

Датафрейм был отсортирован по столбцу SciName.

```python
df_new.drop_duplicates(subset='SciName')
```

Здесь в датафрейме были удалены дубликаты по столбцу SciName.

Целевой признак данной таблицы это CommonNames, т.к. его можно предсказать по основным столбцам.

```python
df_new.to_csv('new_csv.csv')
```

Сохраняем датафрейм в csv файл.

```python
mp.figure(figsize=(27, 6))
sb.countplot(x='CategoryName',data=df_new)
mp.show
```

С помощью библиотек Seaborn и matplotlib.pyplot мы отображаем график по функции countplot по столбцу CategoryName.

```python
mp.figure(figsize=(10, 6))
sb.barplot(y='CategoryName',x='ParkCode',data=df_new)
mp.show
```

С помощью библиотек Seaborn и matplotlib.pyplot мы отображаем график по функции barplot где y это CategoryName а x это ParkCode.

```python
mp.figure(figsize=(10, 6))
sb.boxplot(y='CategoryName',x='ParkCode',data=df_new)
mp.show
```

С помощью библиотек Seaborn и matplotlib.pyplot мы отображаем график по функции boxplot где y это CategoryName а x это ParkCode.

```python
corr = df_new[['References','Observations','Vouchers']].corr()
sb.heatmap(corr, annot=True, cmap='coolwarm')
mp.show
```

Здесь с помощью Seaborn и matplotlib создана heatmap с соотношением столбцов References, Observations и Vouchers.

```python
mp.figure(figsize=(30, 6))
sb.pairplot(df_new)
mp.show()
```

```python
p = sb.PairGrid(df_new)
p.map_lower(sb.scatterplot)
p.map_upper(sb.barplot)
p.map_diag(sb.histplot)
p.add_legend()
mp.show()
```

В отличие от pairplot, pairgrid поддерживает более глубокую настройку и как следствие более подробные данные что видно по графикам сверху.
